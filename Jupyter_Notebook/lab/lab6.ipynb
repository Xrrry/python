```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
import pandas as pd
```

```python
train = pd.read_csv('BirdChicken_TRAIN.tsv',sep='\t')
train
```

```python
train_np = np.array(train)
train_np
```

```python
y = train_np[:1,:500]
z = y.reshape(20,25)
x = y[0]
z.shape
```

```python
me = [np.mean(a[:]) for a in z]
me
```

```python
plt.scatter(range(500),x)
plt.plot(range(500),x)
plt.show()
```

```python
divideIndex = np.array([-0.67, 0, 0.67])
def divide(n):
    if n<divideIndex[0]:
        return 'A'
    elif n<divideIndex[1]:
        return 'B'
    elif n<divideIndex[2]:
        return 'C'
    else:
        return 'D'
```

```python
result = [divide(i) for i in me]
result
```

```python
plt.scatter(range(20), me)
plt.plot(range(20),me)
for i in range(3):
    plt.hlines(divideIndex[i], 0, 20)
for i in range(20):
    plt.annotate(result[i],xy=(i,me[i]),fontsize=14)
plt.show()
```
