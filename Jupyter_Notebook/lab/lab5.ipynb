```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(3)
z = np.random.random(200)*10
z
```

```python
z = z - z.mean()
y = z / z.std()
y
```

```python
y.std()
```

```python
y.mean()
```

```python
y1 = y.reshape(20,10)
```

```python
x = [np.mean(a[:]) for a in y1]
x
```

```python
plt.scatter(range(20),x)
plt.plot(range(20),x)
plt.show()
```

```python
divideIndex = np.array([-0.67, 0, 0.67])
def divide(n):
    if n<divideIndex[0]:
        return 'A'
    elif n<divideIndex[1]:
        return 'B'
    elif n<divideIndex[2]:
        return 'C'
    else:
        return 'D'
```

```python
result = [divide(i) for i in x]
result
```

```python
plt.scatter(range(20), x)
plt.plot(range(20),x)
for i in range(3):
    plt.hlines(divideIndex[i], 0, 20)
for i in range(20):
    plt.annotate(result[i],xy=(i,x[i]),fontsize=14)
plt.show()
```
