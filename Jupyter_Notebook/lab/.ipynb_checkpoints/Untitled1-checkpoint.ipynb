```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
import pandas as pd
```

```python
train = pd.read_csv('Fungi_TEST.tsv',sep='\t')
train
```

```python
train_np = np.array(train)
train_data = train_np[:,1:]
train_target = train_np[:,:1]
```

```python
train_target = train_target.reshape(1,-1)[0]
train_target
```

```python
c = 1
for i in train_target:
    a = int(i)
    if a == c:
        y = train_data[a-1:a,:200]
        z = y.reshape(10,20)
        x = y[0]
        plt.scatter(range(200),x)
        plt.plot(range(200),x)
    else:
        c = a
        plt.show()
```
