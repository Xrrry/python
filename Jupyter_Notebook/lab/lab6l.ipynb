```python
train_np = np.array(train)
train_data = train_np[:,1:]
train_target = train_np[:,:1]
```

```python
divideIndex = np.array([-0.67, 0, 0.67])
def divide(n):
    if n<divideIndex[0]:
        return 'A'
    elif n<divideIndex[1]:
        return 'B'
    elif n<divideIndex[2]:
        return 'C'
    else:
        return 'D'
```

```python
c = 0
for i in train_target:
    a = int(i)
    if a != c:
        c = a
        y = train_data[a-1:a,:400]
        y = y - y.mean()
        y = y / y.std()
        z = y.reshape(20,20)
        me = [np.mean(a[:]) for a in z]
        result = [divide(i) for i in me]
        x = y[0]
        plt.scatter(range(20), me)
        plt.plot(range(20),me)
        for i in range(3):
            plt.hlines(divideIndex[i], 0, 20)
        for i in range(20):
            plt.annotate(result[i],xy=(i,me[i]),fontsize=14)
        plt.show()
```
