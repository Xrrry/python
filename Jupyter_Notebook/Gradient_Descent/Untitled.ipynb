```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plot_x = np.linspace(-1,6,141)
plot_y = (plot_x-2.5)**2+2
```

```python
plt.plot(plot_x,plot_y)
plt.show()
```

```python
def dj(theta):
    return 2*(theta-2.5)
```

```python
def j(theta):
    return (theta-2.5)**2+2
```

```python
theta = 0.0
epsilon = 1e-8
eta = 0.1
while True:
    gradient = dj(theta)
    last_theta = theta
    theta = theta - gradient * eta
    if abs(j(theta) - j(last_theta)) < epsilon:
        break
print(theta)
print(j(theta))
```

```python
theta = 0.0
eta = 0.001
theta_history = [theta]
while True:
    gradient = dj(theta)
    last_theta = theta
    theta = theta - gradient * eta
    theta_history.append(theta)
    
    if abs(j(theta) - j(last_theta)) < epsilon:
        break
        
plt.plot(plot_x, j(plot_x))
plt.plot(np.array(theta_history), j(np.array(theta_history)), color='r', marker='+')
plt.show()
len(theta_history)
```
